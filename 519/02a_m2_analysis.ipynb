```python
import xarray as xa
```

```python
from muscat2ta.m2mnlpf import M2MultiNightLPF, read_reduced_m2
from muscat2ph.catalog import get_toi
```

```python
toi = get_toi(519.01)
```

```python
lpf = M2MultiNightLPF('toi00519.01', True, radius_ratio='chromatic', klims=(0.23, 0.35))
```

```python
lpf.set_prior('tc', 'NP', lpf.times[0].mean()+lpf.tref,  0.01)
lpf.set_prior('p',  'NP', toi.period[0], 5*toi.period[1])
```

```python
lpf.set_radius_ratio_prior(0.23, 0.35)
```

```python
lpf.optimize_global(15000, npop=400)
```

```python
lpf.plot_light_curves(figsize=(13,9));
```

```python
lpf.sample_mcmc(5000, thin=25, repeats=10, save=True)
```

```python
lpf.plot_chains()
```

```python
lpf.save()
```
